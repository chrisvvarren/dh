## But what's this, this second dataset has phone numbers...and so does the first...

## It just takes one line of code to match the first dataset to the second

```python
df.set_index('phone1').join(other.set_index('phone1'))
```

## Why would someone want to do that?

Well, now they can very easily find everyone in, say, PA with an income above, say, $80,000 

```python
joined = df.set_index('phone1').join(other.set_index('phone1'))
PA = joined.loc[joined['state'] == 'PA']
PA.loc[PA['Income'] > 80000]
```

## and email them...

```python
PA.loc[PA['Income'] > 80000]['email']
```

## Device fingerprinting or browser fingerprinting 

According to [AmIunique.org](https://amiunique.org/faq), browser fingerprinting is

> the systematic collection of information about a remote device, for identification purposes. Client-side scripting languages allow the development of procedures to collect very rich fingerprints: browser and operating system type and version, screen resolution, architecture type, lists of fonts, plugins, microphone, camera, etc.

Collected data frequently includes:

- the User agent header
- the Accept header
- the Connection header
- the Encoding header
- the Language header
- the Upgrade Insecure Requests header
- the Referer header
- the Cache-Control header
- the BuildId of the browser
- the list of plugins
- the platform
- the cookies preferences (allowed or not)
- the Do Not Track preferences (yes, no or not communicated)
- the timezone
- the screen resolution and its color depth
- the use of local storage
- the use of session storage
- a picture rendered with the HTML Canvas element
- a picture rendered with WebGL
- Supported Audio formats
- Supported Video formats
- the presence of AdBlock
- the list of fonts

### How I Made the Fake Data

```python
phones = df['phone1'].copy()
random.shuffle(phones)
data = {}
for i in range(0,500):
    data[i] = [random.randrange(35000,600000,100), random.random(), phones[i]]
    
other = pd.DataFrame.from_dict(data, orient='index', dtype=None, columns=['Income', 'Gun Index', 'phone1'])
other.to_csv('the_join_example2.csv', index=None)
```

```python
addresses = df['address'].copy()
random.shuffle(addresses)
data = {}
for i in range(0,500):
    data[i] = [random.randrange(5,13,1), addresses[i]]
third = pd.DataFrame.from_dict(data, orient='index', dtype=None, columns=['Age', 'Address'])
browser = pd.read_csv('fake_browser_data.csv')
browser = browser.dropna()
agent = browser['User agent'].copy().reset_index(drop=True)
vendor = browser['Mac vendor'].copy().reset_index(drop=True)
when = browser['Discovered when'].copy().reset_index(drop=True)
random.shuffle(agent)
random.shuffle(vendor)
random.shuffle(when)
data = {}
for i in range(0,500):
    j = random.randint(0, 47)
    data[i] = [agent[j], vendor[j], when[j]]
browser = pd.DataFrame.from_dict(data, orient='index', dtype=None, columns=['User agent', 'Mac vendor', 'Discovered when'])
browser = pd.concat([third,browser], axis=1)
browser.to_csv('the_join_example3.csv', index=None)
```
