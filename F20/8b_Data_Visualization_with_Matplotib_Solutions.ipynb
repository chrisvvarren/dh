Using the `matplotlib`, `pandas,` and the two files, `locations.csv` and `places.csv`, try to re-create the image below. 

````python
Image('matplotlib_goal.png')
````

````python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(15,8), sharey=True)
places = pd.read_csv('places.csv', index_col=0)
places['Names minus Places'].sort_values(ascending=False).plot(kind='bar', color='red', ax=ax1).set_title('People vs. Places', fontweight='bold')
ax1.axhline(0, color='k')
locations = pd.read_csv('locations.csv', index_col=0)
locations['Foreign vs. Domestic Places'].sort_values(ascending=False).plot(kind='bar', color='red', ax=ax2).set_title('Foreign Places vs. Domestic Places', fontweight='bold')
ax2.axhline(0, color='k')
ax1.set(ylabel='Difference in mentions per 100 words')
ax1.text(6, .8, 'People', fontsize=15)
ax1.text(20, -1, 'Places', fontsize=15)
ax2.text(.1, .6, 'Non-British Isles Places', fontsize=15)
ax2.text(6, -1, 'British Isles Places', fontsize=15)
plt.show()
#fig.tight_layout()
#fig.savefig('matplotlib_goal.png')
````
